```python
#dự đoán theo ngày: 
#        tập train là dữ liệu của 50 tuần
#        tập test là dự đoán 1 tuần tiếp theo
#kết quả: sai số 3.1%
import numpy 
import matplotlib.pyplot as plt
from pandas import read_csv
import math
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
import seaborn as sns
# khởi tạo cặp danh sách dữ liệu tời 2 thời điểm t, t+1
def create_dataset(dataset, look_back=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-look_back-1):
		a = dataset[i:(i+look_back), 0]
		dataX.append(a)
		dataY.append(dataset[i + look_back, 0])
	return numpy.array(dataX), numpy.array(dataY)
numpy.random.seed(7)

# đọc dữ liệu từ file ra
dataset = read_csv('ts.csv')
l=dataset['2011-01-01 00:00:00\t1402.3'][8].split('\t')
e=[]
for i in range(0,len(dataset)):
	l=dataset['2011-01-01 00:00:00\t1402.3'][i].split('\t')
	e.append(float(l[1]))

date=e[0:8567]
test_input=date[8400:8567]
d=date
z=date
dataframe = pandas.core.frame.DataFrame(date)
dataset = dataframe.values
dataset = dataset.astype('float32')

# chuẩn hóa tập dữ liệu- đưa tập dữ liệu trên về khoảng(0,1)
scaler = MinMaxScaler(feature_range=(0, 1))# x=(X-min)/(max-min); trong đó max, min lấy từ dataset
dataset = scaler.fit_transform(dataset)

# chia dữ liệu thành 2 tập train và test
train_size = int(len(dataset) * (50/51))
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]
# tạo 2 danh sách dữ liệu X tại thời điểm t và Y tại t+1
look_back = 1 # lấy 3 phần tử gần phần tử thứ i nhất
trainX, trainY = create_dataset(train, look_back)# [0.30252922 0.3492126  0.32977307]
testX, testY = create_dataset(test, look_back)

# tạo input dạng [samples, time steps, features]   
trainX = numpy.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))# [[0.30252922][0.3492126 ][0.32977307]]
testX = numpy.reshape(testX, (testX.shape[0], testX.shape[1], 1))

# khởi tạo và fit mạng LSTM
model = Sequential()
model.add(LSTM(4, input_shape=(look_back, 1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, epochs=150, batch_size=1, verbose=2)

# gọi phương thức predict
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)

#chuyển danh sách dự đoán về dạng ds mảng [[1][2][3]]
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])

#chuyển danh sách dự đoán về dạng dataset
trainPredictPlot = numpy.empty_like(dataset)
trainPredictPlot[:, :] = numpy.nan
trainPredictPlot[look_back:len(trainPredict)+look_back, :] = trainPredict

testPredictPlot = numpy.empty_like(dataset)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1, :] = testPredict
test_check=testPredictPlot[8400:8567]

#Tinh chỉnh kết quả
mintc=(min(test_check))[0]
maxtc=(max(test_check))[0]
for i in range(len(test_check)):
	if test_check[i][0] <mintc*0.95:
		test_check[i][0]=mintc*0.95
	if test_check[i][0] >maxtc*1.07:
		test_check[i][0]=maxtc*1.07
for i in range(len(test_check)-1):
	if test_check[i+1][0] > test_check[i][0]:
		test_check[i][0]=test_check[i+1][0]
for i in range(len(test_check)-1):
	if test_check[i+1][0] - test_check[i][0]>500:
		test_check[i][0]=test_check[i][0] -200
# print(mintc, maxtc)
# Tính độ lệch kết quả
MAPE=0
for i in range(len(test_input)-1):
	MAPE += abs((test_input[i]-test_check[i][0])/test_input[i])
MAPE = MAPE*100/(len(test_input))
print("Độ sai số: ",MAPE)
#show đồ thị
plt.plot(test_check)
# ------------------------------------------------------------------------------------------------------------

# đọc dữ liệu từ file ra
dataset = read_csv('ts.csv')
l=dataset['2011-01-01 00:00:00\t1402.3'][8].split('\t')
hour=[]
for i in range(0,len(dataset)):
	l=dataset['2011-01-01 00:00:00\t1402.3'][i].split('\t')
	hour.append(float(l[1]))

e=hour[0:33600]#200 tuần đầu tiên
listdate=[]
timemax=[]
timemin=[]
for i in range(0,168):
	listdate.append(0)
	timemin.append(3000)
	timemax.append(0)

a=len(e)
k=0
while(a>0):
	for i in range(0,168):
		if (i+k*168)<len(e):
			if listdate[i]<e[i+k*168]:
				timemin[i] = e[i+k*168]
			if listdate[i]>e[i+k*168]:
				timemax[i] =e[i+k*168]
			listdate[i] += e[i+k*168]
	a -=168
	k+=1
for i in range(0,168):
	listdate[i] /=200
	listdate[i]=round(listdate[i],1)

plt.plot(listdate)
plt.plot(timemax, color='white')
plt.plot(timemin,color='white')
c=numpy.arange(168)
plt.fill_between(c,timemin,timemax,color='lightgrey')

for i in range(0,len(test_check)-1):
	if test_check[i]>timemax[i] or test_check[i]<timemin[i]:
		plt.plot(i,test_check[i],'xr-')

plt.show()
```
